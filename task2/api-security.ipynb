```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('./api-security-dataset.csv')
```

```python
df
```

```python
df.info()
```

```python
df = df.iloc[:, 2:12]
```

```python
X = df.iloc[:, 0:9]
```

```python
y = df.iloc[:, 9]
```

```python
X.head()
```

```python
y
```

```python
X['ip_type'] = X.ip_type.astype('category').cat.codes
```

```python
X['source'] = X.source.astype('category').cat.codes
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 21, test_size=0.30)
```

```python
print(type(X), type(y))
print(X_train.shape, y_train.shape)
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc = DecisionTreeClassifier(max_depth = 3)
```

```python
dtc.fit(X_train, y_train)
```

```python
y_pred = dtc.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f"Decision tree Accuracy: {accuracy:.4f}")
```

```python
features = pd.DataFrame(dtc.feature_importances_, index = X.columns)
features
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(
    n_estimators = 10,
    criterion = "gini",
    max_depth = 9,
    max_samples=0.8,
    random_state=21)
```

```python
rf.fit(X_train, y_train)
y_rf_pred = rf.predict(X_test)
```

```python
print(confusion_matrix(y_test, y_rf_pred))
```

```python
print(classification_report(y_test, y_rf_pred))
```

```python
accuracy_rf = accuracy_score(y_test, y_rf_pred)
print(f"Random Forest Accuracy: {accuracy_rf:.4f}")
```
